**4.- Indexación**
Nos permite acceder o extraer porciones de un DataFrame o de una Serie y es una operación muy pero muy común al momento de procesar y analizar datos tabulares.

**4.1.- Extracción de una o multiples columnas**
En este caso usamos la sintaxis df['nombre_columna'] o df[[listado_de_columnas]].
Por ejemplo, extraigamos solo una columna

```
df_csv['song_title']
```

```
#Extraemos dos comlumnas
df_csv[['song_title','artist']]
```

4.2. Extraer un rango de filas
El metódo "iloc()": nos permite extarer parte de las filas del DataFrame o Serie.
Por ejemplo extraer las primeras 10 filas:

```
df_csv.iloc[0:10,:] #tambien se puede usar [:10]
```

Uso de ":" como segundo argumento indica que tomaremos todas las columnas.
Si queremos extraer las primeras 10 filas pero únicamente las columnas desde el "indice 1" al "indice 4"

```
df_csv.iloc[0:10,1:4]
```

# 4.3 Filtrado de Filas y Columnas
Podemos hacer uso de **condicionales** para extraer porciones de un DataFrame.
Por ejemplo, extraigamos únicamente la popularidad arriba de 50. Podemos utilizar el metódo "loc"

```
df_csv.loc[df_csv['popularity'] > 50]
```

Y en el argumento de "loc" también podemos inlcluir las columanas a extraer. Por ejemplo, usemos el codicional anterior y extraigamos ademas las columnas composer

```
df_csv.loc[df_csv['popularity'] > 50, ['label','composer']]
```

# 5. Manipulando el contenido de un DataFrame
Se pueden usar herramientas para generar nuevas tablas manipulando el contenido de la tabla (DataFrame) original.

# 5.1 ordenamiento de DataFrames
Podemos **ordenar** las filas de una tabla con respecto a un cierto criterio.
Por ejemplo, ordenaremos las filas usando como criterio la popularidad de menor a mayor

```
df_csv.sort_values(by='popularity')
```

Y para ordenar de manera descendente lo podemos hacer usando el argumento "ascending=False":

```
df_csv.sort_values(by='popularity', ascending=False)
```

Y podemos utilizar múltiples criterios de ordenamiento. Por ejemplo ordenemos de manera ascendente por popularidad y duration

```
df_csv.sort_values(by=['popularity','duration'])
```

Y esta lógica también se puede aplicar para ordenar alfabéticamente usando como criterio variables categóricas.
Por ejemplo, ordenaremos el DataFrame por lenguaje de A-Z

```
df_csv.sort_values(by='language', ascending=False)
```

# 5.2 Renombrar columnas
Podemos cambiar el nombre de una o varias columnas usando el método "rename".
Como argumento de este método usamos un deiccionario que contendrá el nombre original y el nuevo nombre para cada columna que queramos renombrar

```
df_csv.rename(columns={'song_title':'Nombre_Cancion'})
```

# 6. Agrupaciones
El método "GroupBy" resulta muy útil cuando queremos usar una variable categórica para generar agrupaciones.
La sintaxis básica es:
**DataFrame -> groupby -> función de agregación

Donde la función de agregación permite, como su nombre lo indica, agregar los valores de cada agrupación generada(por ejemplo sumándolos, promediándolos, etc.)
Por ejemplo, generaremos agrupaciones usando como cvriterio el genero y por cada genero calculemos el promedio de popularidad.

```
df_csv.groupby('genre')['popularity'].mean()
```

Y podemos organizar lo anterior por ejemplo de manera descendente por los valores promedios de popularidad:

```
df_popularidad = df_csv.groupby('genre')['popularity'].mean().sort_values(ascending=False)
```

```
df_popularidad
```
